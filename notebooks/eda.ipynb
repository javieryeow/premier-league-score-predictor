```python
import pandas as pd
df = pd.read_csv("../data/PremierLeague.csv")
df.head()
df.info()
df.describe()
```

```python
# analyse missing values
missing = df.isnull().sum()
missing[missing > 0].sort_values(ascending=False)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# examining home and away goals distributions

sns.histplot(df["FullTimeHomeTeamGoals"], bins=10, kde=False)
plt.title("Distribution of Home Goals")

```

```python
sns.histplot(df["FullTimeAwayTeamGoals"], bins=10, kde=False)
plt.title("Distribution of Away Goals")
```

```python
# examining match outcomes

sns.countplot(x="FullTimeResult", data=df)
plt.title("Match Outcomes (H/D/A)")
```

```python
# scoring patterns over time

# convert date to datetime
df["Date"] = pd.to_datetime(df["Date"], errors="coerce")
df["Year"] = df["Date"].dt.year

df.groupby("Year")[["FullTimeHomeTeamGoals", "FullTimeAwayTeamGoals"]].mean().plot()
plt.title("Average Goals per Year")

```

```python
# average performance by teams in terms of goals scored both home and away

home_avg = df.groupby("HomeTeam")["FullTimeHomeTeamGoals"].mean().sort_values(ascending=False)
away_avg = df.groupby("AwayTeam")["FullTimeAwayTeamGoals"].mean().sort_values(ascending=False)
team_perf = pd.DataFrame({
    "Avg Home Goals": home_avg,
    "Avg Away Goals": away_avg
})
team_perf.plot(kind="bar", figsize=(16, 6))
plt.title("Average Goals Scored Per Team (Home vs Away)")

```

```python
# correlation between stats 

numerics = df.select_dtypes(include=["float64", "int64"])
correlation = numerics.corr()
print(correlation)
```
